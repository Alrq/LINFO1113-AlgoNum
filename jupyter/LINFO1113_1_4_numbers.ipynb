# Norm example

```python
from math import sqrt
def norm(x):
    squaredSum = 0;
    for x_i in x:
        squaredSum += x_i*x_i
    return sqrt(squaredSum)
```

```python
print("%e" % norm([1e100, 1e100]))
print("%e" % norm([1e200, 1e200]))
```

```python
def norm_improved(x):
    maxElement = max(x)
    squaredSum = 0
    for x_i in x:
        #scale x_i
        x_i_scaled = x_i/maxElement
        squaredSum += x_i_scaled*x_i_scaled

    #donâ€™t forget to unscale the result
    return sqrt(squaredSum) * maxElement

```

```python
print("%e" % norm_improved([1e100, 1e100]))
print("%e" % norm_improved([1e200, 1e200]))
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-1e-6, 1e-6, 1e-8, np.float32)
y = ((np.exp(x)-1) / x) - 1

plt.figure(figsize=(10, 10))
plt.plot(x, y)
```
